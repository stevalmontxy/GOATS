This notebook is just a convenience diagnostic feature to check the contents of the portfolio from the device that is running the scripts

import required stuff

```python
import shelve, os, sys, dotenv
sys.path.append(os.path.abspath(".."))
shelf_path = os.path.abspath("../scripts/goats_shelf")
from goats.core.core_objects import Portfolio, Stock, Option, Position, Account
from goats.core.core_objects import Order, LimitOrder, ScheduledOrder
from goats.broker.broker import Broker
from goats.broker.alpaca_broker import AlpacaBroker

dotenv.load_dotenv()
trading_mode = os.getenv("TRADING_MODE")
paper = (trading_mode == "paper") # set boolean
if paper:
    api_key = os.getenv("ALPACA_API_KEY_PAPER")
    secret_key = os.getenv("ALPACA_SECRET_KEY_PAPER")
else:
    api_key = os.getenv("ALPACA_API_KEY_LIVE")
    secret_key = os.getenv("ALPACA_SECRET_KEY_LIVE")
```

load port from shelf

```python
with shelve.open(shelf_path) as db:
    port: Portfolio = db.get('portfolio') # using db.get will return None if not found instead of error

port
```

```python
port.pseudo_orders
```

Add broker and update portfolio

```python
alpaca_broker = AlpacaBroker(api_key=api_key, secret_key=secret_key, paper=paper)
# strat = DemoStrat(alpaca_broker, port)
port.set_broker(alpaca_broker)
port.update_portfolio()

# NOTE: this does not affect the shelf port object. would need to resave if you want to 
port
```

Wipe broker and save to shelf

```python
wipe_broker = True
if wipe_broker:
    port.set_broker(None)

with shelve.open(shelf_path) as db:
    db.update({'portfolio': port})
```
