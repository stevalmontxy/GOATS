this file is for executing a backtest. it's easier to do in a jupyter notebook

Imports

```python
# Standard Imports
import datetime as dt
import sys
import os
from dotenv import load_dotenv

# Third Party Imports
import pandas as pd
# import plotly.graph_objects as go

# Local Imports
sys.path.append(os.path.abspath(".."))
from goats.core.core_objects import Portfolio, Stock, Option, Position, Account
from goats.core.core_objects import Order, LimitOrder, ScheduledOrder
from goats.core.strategy import Strategy
from goats.core.backtest import Backtest
from goats.broker.broker import Broker
from goats.broker.bt_broker import BTBroker
```

Import data

```python
# import option data
options_df = pd.read_csv("../data/aapl_30x_202307.txt", delimiter=", ")
options_df = options_df.drop(options_df.columns[[0, 6, 8,9,10,11,12, 13,14, 16,17,18, 20,21,22, 24,25,26,27,28, 29]], axis=1) # dont forget to acct for 0 ind
options_df["[QUOTE_DATE]"] = pd.to_datetime(options_df["[QUOTE_DATE]"])
options_df["[EXPIRE_DATE]"] = pd.to_datetime(options_df["[EXPIRE_DATE]"])
# import underlying data
file_dir = "../data/AAPL.USUSD_Candlestick_30_M_BID_02.01.2023-29.12.2023FORMATTED.csv"
underly_df = pd.read_csv(file_dir, parse_dates=['Datetime'], index_col='Datetime')
underly_df = underly_df[underly_df['Volume']!=0] # remove data with zero volume
```

Setup

```python
initial_cash = 10000
broker = BTBroker(options_df, underly_df, initial_cash)
port = Portfolio(broker, initial_cash)
strat = Strategy(broker, port)
# strat's broker is same instance as port's broker when defined this way

bt = Backtest(strat, options_df, underly_df)
```

```python
# Import and swap out strategy
from goats.core.strategy import DemoStrat

# data = read_data(csv_file)
# bt.import_data(data)

demo_strat = DemoStrat(broker, port)
bt.set_strat(demo_strat)
```

Execute

```python
start_date = '2023-07-01' # format 'YYYY-MM-DD'
end_date = '2023-07-29'
bt.run(start_date, end_date)
```
